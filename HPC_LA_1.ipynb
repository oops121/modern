```python
%%writefile bfs_dfs.cpp

#include <iostream>
#include <vector>
#include <queue>
#include <fstream>
#include <omp.h>
#include <stack>

using namespace std;

class Graph {
public:
    int V;
    vector<vector<int>> adj;

    Graph(int V) {
        this->V = V;
        adj.resize(V);
    }

    void addEdge(int u, int v) {
        adj[u].push_back(v);
        adj[v].push_back(u);
    }


    // ---------------- SEQ BFS ----------------
    void seqBFS(int start) {
        vector<char> visited(V, 0);
        queue<int> q;

        visited[start] = 1;
        q.push(start);

        while (!q.empty()) {
            int node = q.front();
            q.pop();

            for (int n : adj[node]) {
                if (!visited[n]) {
                    visited[n] = 1;
                    q.push(n);
                }
            }
        }
    }

    // ---------------- PAR BFS ----------------
    void parBFS(int start) {
        vector<char> visited(V, 0);
        vector<int> frontier{start};

        visited[start] = 1;

        while (!frontier.empty()) {
            vector<int> next;

            #pragma omp parallel
            {
                vector<int> local;

                #pragma omp for schedule(dynamic)

                for (int i = 0; i < frontier.size(); i++) {

                    int node = frontier[i];

                    for (int n : adj[node]) {

                        if (!visited[n]) {

                            bool added = false;

                            #pragma omp critical
                            {
                                if (!visited[n]) {
                                    visited[n] = 1;
                                    added = true;
                                }
                            }

                        if (added)
                              local.push_back(n);


                        }


                    }
                }


                #pragma omp critical
                next.insert(next.end(), local.begin(), local.end());

            }

            frontier.swap(next);

        }
    }

    // ---------------- SEQ DFS ----------------
    void seqDFS(int start) {
        vector<char> visited(V, 0);
        stack<int> st;
        st.push(start);

        while (!st.empty()) {
            int node = st.top();
            st.pop();

            if (visited[node]) {
                    continue;
            }

            visited[node] = 1;

            for (int n : adj[node]) {
                if (!visited[n]) {
                    st.push(n);
                  }
              }

        }  // end while

    }

    // ---------------- PAR DFS ----------------
    void parDFS(int start) {
        vector<char> visited(V, 0);
        vector<int> curr{start};

        while (!curr.empty()) {
            vector<int> next;

            #pragma omp parallel
            {
                vector<int> local;

                #pragma omp for schedule(dynamic)

                for (int i = 0; i < curr.size(); i++) {
                    int node = curr[i];

                    bool process = false;

                    if (!visited[node]) {

                        #pragma omp critical

                        {
                            if (!visited[node]) {
                                visited[node] = 1;
                                process = true;
                            }
                        }
                    }

                    if (process) {

                        for (int n : adj[node]) {
                            if (!visited[n])
                                local.push_back(n);
                        }
                    }


                }

                #pragma omp critical

                next.insert(next.end(), local.begin(), local.end());

            }

            curr.swap(next);


        }


    }


};

// ---------------- MAIN ----------------
int main() {

    ofstream file("result1.txt");

    file << "N,BFS_SEQ,BFS_PAR,BFS_SPEEDUP,BFS_EFF,BFS_COST,"
            "DFS_SEQ,DFS_PAR,DFS_SPEEDUP,DFS_EFF,DFS_COST\n";

    //  FORCE THREADS
    int cores = 4;
    omp_set_num_threads(cores);

    cout << "Using " << cores << " threads (limited for crossover detection)\n";

    for (int N = 100; N <= 3000; N += 500) {

        Graph g(N);

        //  scalable density
        for (int i = 0; i < N; i++) {
            for (int j = i + 1; j < min(N, i + (N / 20)); j++) {
                g.addEdge(i, j);
            }
        }

        double t1, t2;

        // ================= BFS =================
        t1 = omp_get_wtime();
        g.seqBFS(0);
        t2 = omp_get_wtime();
        double bfs_seq = t2 - t1;

        t1 = omp_get_wtime();
        g.parBFS(0);
        t2 = omp_get_wtime();
        double bfs_par = t2 - t1;

        double bfs_speed = bfs_seq / bfs_par;
        double bfs_eff = bfs_speed / cores;
        double bfs_cost = bfs_par * cores;

        // ================= DFS =================
        t1 = omp_get_wtime();
        g.seqDFS(0);
        t2 = omp_get_wtime();
        double dfs_seq = t2 - t1;

        t1 = omp_get_wtime();
        g.parDFS(0);
        t2 = omp_get_wtime();
        double dfs_par = t2 - t1;

        double dfs_speed = dfs_seq / dfs_par;
        double dfs_eff = dfs_speed / cores;
        double dfs_cost = dfs_par * cores;

        // ================= PRINT =================
        cout << "\n====================================\n";
        cout << "N = " << N << endl;

        cout << "\n--- BFS ---\n";
        cout << "Sequential Time : " << bfs_seq << endl;
        cout << "Parallel Time   : " << bfs_par << endl;
        cout << "Speedup         : " << bfs_speed << endl;
        cout << "Efficiency      : " << bfs_eff << endl;
        cout << "Cost            : " << bfs_cost << endl;

        cout << "\n--- DFS ---\n";
        cout << "Sequential Time : " << dfs_seq << endl;
        cout << "Parallel Time   : " << dfs_par << endl;
        cout << "Speedup         : " << dfs_speed << endl;
        cout << "Efficiency      : " << dfs_eff << endl;
        cout << "Cost            : " << dfs_cost << endl;

        // ================= FILE =================
        file << N << ","
             << bfs_seq << "," << bfs_par << "," << bfs_speed << "," << bfs_eff << "," << bfs_cost << ","
             << dfs_seq << "," << dfs_par << "," << dfs_speed << "," << dfs_eff << "," << dfs_cost << "\n";



    }

    file.close();
    cout << "\nSaved to result1.txt\n";
}
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv("result1.txt")

N = df["N"].values

# ================= BFS =================
bfs_seq = df["BFS_SEQ"].values
bfs_par = df["BFS_PAR"].values
bfs_speed = df["BFS_SPEEDUP"].values

# ================= DFS =================
dfs_seq = df["DFS_SEQ"].values
dfs_par = df["DFS_PAR"].values
dfs_speed = df["DFS_SPEEDUP"].values

#  FIX: correct cores
CORES = 4

bfs_eff = bfs_speed / CORES
dfs_eff = dfs_speed / CORES

# ================= BFS TIME GRAPH =================
plt.figure()

plt.plot(N, bfs_seq, marker='o', label="BFS Sequential")
plt.plot(N, bfs_par, marker='o', label="BFS Parallel")

plt.xlabel("Input Size N")
plt.ylabel("Time (seconds)")
plt.title("BFS: Sequential vs Parallel")

#  SHOW ALL VALUES
plt.xticks(N, rotation=90)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ================= DFS TIME GRAPH =================
plt.figure()

plt.plot(N, dfs_seq, marker='o', label="DFS Sequential")
plt.plot(N, dfs_par, marker='o', label="DFS Parallel")

plt.xlabel("Input Size N")
plt.ylabel("Time (seconds)")
plt.title("DFS: Sequential vs Parallel")

plt.xticks(N, rotation=90)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ================= BFS SPEEDUP =================
plt.figure()

plt.plot(N, bfs_speed, marker='o', label="BFS Speedup")

#  crossover reference
plt.axhline(y=1, linestyle='--')

plt.xlabel("Input Size N")
plt.ylabel("Speedup")
plt.title("BFS Speedup")

plt.xticks(N, rotation=90)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ================= DFS SPEEDUP =================
plt.figure()

plt.plot(N, dfs_speed, marker='o', label="DFS Speedup")

plt.axhline(y=1, linestyle='--')

plt.xlabel("Input Size N")
plt.ylabel("Speedup")
plt.title("DFS Speedup")

plt.xticks(N, rotation=90)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ================= BFS EFFICIENCY =================
plt.figure()

plt.plot(N, bfs_eff, marker='o', label="BFS Efficiency")

plt.axhline(y=1, linestyle='--')

plt.xlabel("Input Size N")
plt.ylabel("Efficiency")
plt.title("BFS Efficiency")

plt.xticks(N, rotation=90)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ================= DFS EFFICIENCY =================
plt.figure()

plt.plot(N, dfs_eff, marker='o', label="DFS Efficiency")

plt.axhline(y=1, linestyle='--')

plt.xlabel("Input Size N")
plt.ylabel("Efficiency")
plt.title("DFS Efficiency")

plt.xticks(N, rotation=90)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ================= COST GRAPH =================
bfs_cost = df["BFS_COST"].values
dfs_cost = df["DFS_COST"].values

plt.figure()

plt.plot(N, bfs_cost, marker='o', label="BFS Cost")
plt.plot(N, dfs_cost, marker='o', label="DFS Cost")

plt.xlabel("Input Size N")
plt.ylabel("Cost")
plt.title("Cost Analysis")

plt.xticks(N, rotation=90)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# ================= CROSSOVER DETECTION =================
plt.figure()

plt.plot(N, bfs_seq, marker='o', label="BFS Seq")
plt.plot(N, bfs_par, marker='o', label="BFS Par")

# crossover reference
plt.axhline(y=0)

plt.xlabel("Input Size N")
plt.ylabel("Time")
plt.title("Crossover Detection BFS")

plt.xticks(N, rotation=90)

plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

#  FIND CROSSOVER POINT
idx = np.argmin(np.abs(bfs_seq - bfs_par))
print("Crossover near N =", N[idx])
```
